```python
import pandas as pd

# Load Excel
file_path = '/Users/yoshifumihanada/Downloads/シノッカンピソㇱ8択問題ベース.xlsx'
df = pd.read_excel(file_path, sheet_name=0)
df.drop(['Unnamed: 0'], axis=1, inplace=True)

# Show DataFrame
print(df.info())
```

```python
# ilocがなぜか使えないので、列削除で対応

ain_quiz = df.drop(df.columns[[0, 1, 4, 6, 8, 10, 12, 14, 16, 18, 19]], axis=1)
ain_quiz.head()

output_path = './quiz_reordered.txt'

txt_buffer = []

for i in range(len(ain_quiz)):
    prefix = "'ACT03_{0:03d}".format(i)
    question = prefix + "_quiz': '" + ain_quiz.iloc[i,0] + "'," 

    txt_buffer.append(question)
    print(question)

    for j in range(1,9):
        op_desc = ain_quiz.iloc[i,j] if ain_quiz.iloc[i,j] != 'nan' else ''
        op = prefix + "_op{0}': '{1}',".format(j, op_desc)

        txt_buffer.append(op)
        print(op)

    txt_buffer.append('')
    print('='*20)

with open(output_path, 'w') as file:
    for line in txt_buffer:
        file.write(line + '\n')
```
