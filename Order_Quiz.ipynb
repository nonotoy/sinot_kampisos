```python
import pandas as pd

# Load Excel
file_path = '/Users/yoshifumihanada/Downloads/「シノッカンピソㇱ」設問 データ完成版.xlsx'
df = pd.read_excel(file_path, sheet_name=0)
#df.drop(['Unnamed: 0'], axis=1, inplace=True)

# Show DataFrame
print(df.info())
```

```python
output_path = './quiz_reordered.txt'

txt_buffer = []

genre = {
    '生物': 'animals',
    '植物': 'plants',
    '自然': 'nature',
    '場所': 'places', 
    'ひと': 'human',
    '人体': 'body',
    '動作': 'action',
    '性質': 'property',
    '道具': 'tool',
    '食べ物': 'food',
    '時間': 'time',
    '数': 'number',
    '空間': 'space',
    '上級': 'advanced',
    '親子教室': 'beginner',
}

level = {
    '簡': 'easy',
    '難': 'hard',
}

prefix_list=[]

for i in range(len(df)):

    if i == 0:
        q = 0
    elif df.iloc[i,0] != df.iloc[i-1,0] or df.iloc[i,1] != df.iloc[i-1,1]:
        q = 0
    else:
        q += 1
        
    prefix = "'ACT02_{0}_{1}_{2}".format(genre[df.iloc[i,0]],level[df.iloc[i,1]],format(q, '03'))

    #prefix_ = "'ACT02_{0}_{1}".format(genre[df.iloc[i,0]],level[df.iloc[i,1]])
    #prefix_list.append(prefix_)

    question = prefix + "_quiz': '" + df.iloc[i,2] + "'," 

    txt_buffer.append(question)
    print(question)

    for j in [3,5,7,9]:
        if not pd.isnull(df.iloc[i,j]):
            op_desc = df.iloc[i,j]
            op = prefix + "_op{0}': '{1}',".format(int((j-1)/2), op_desc)

            txt_buffer.append(op)
            print(op)

    txt_buffer.append('')

    print('')

print(set(prefix_list))
```
